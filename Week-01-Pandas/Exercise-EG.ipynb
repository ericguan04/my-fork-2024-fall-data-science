## Exercise 2 - Filtering

Return the following subsets of the dataframe.

1. How many listings are there with a price less than 100? 


2. Find how many listings there are in just Brooklyn.


3. Find how many listings there are in Brooklyn with a price less than 100.


4. Using `.isin()` select anyone that has the host name of Michael, David, John, and Daniel.


5. Create a new column called `adjusted_price` that has $100 added to every listing in Williamsburg.  The prices for all other listings should be the same as the were before. 


6. What % of the rooms are private, and what % of the rooms are shared.  
    * Hint, use `.value_counts()`


```python
# 1. How many listings are there with a price less than 100? 
df.head()

select_condition = df["price"] < 100
df[select_condition] # alternative is to explicitly write select_condition in brackets

len(df[select_condition])
```

```python
# 2. Make a new DataFrame of listings in Brooklyn named `df_bk` 
# and find how many listings in just Brooklyn.
df.head()

select_condition = df["neighbourhood_group"] == "Brooklyn"
df_bk = df[select_condition].copy() # alternative is to explicitly write select_condition in brackets; remember to add .copy()

df_bk
len(df_bk)

```

```python
# 3. Find how many listings there are in Brooklyn with a price less than 100.
df_bk.head()

select_condition = df_bk["price"] < 100
df_bk[select_condition]  # alternative is to explicitly write select_condition in brackets

len(df_bk[select_condition])
```

```python
# 4. Using `.isin()` select anyone that has the host name of Michael, David, John, and Daniel.
# How many total are there that have those names

df.head()
host = ["Michael", "David", "John", "Daniel"]
condition = df.host_name.isin(host)  # alternative is to explicitly write ["names", ...] in the isin() function

df[condition]["host_name"].value_counts() # counts total of each category in host_name
```

```python
# 5. Create a new column called `adjusted_price` that has $100 added to every listing in Williamsburg.  
# The prices for all other listings should be the same as the were before. 

def adj_price(row):
    if row["neighbourhood"] == "Williamsburg":
        return row["price"] + 100
    else:
        return row["price"] 

# .apply() takes df as input for the specified function
df["adjusted_price"] = df.apply(adj_price, axis=1) # axis=1 applies the function row-wise
df
```

```python
# 6. What % of the rooms are private, and what % of the rooms are shared.  

df.head()

dfp = df["room_type"].value_counts(normalize=True) # normalize parameters gives us probabilities
dfp
```

```python
dfp[["Private room", "Shared room"]] # select only relevant categories
```

# Exercise 3 - Grouping

1. Using `groupby`, count how many listings are in each neighbourhood_group.


2. Using `groupby`, find the mean price for each of the neighbourhood_groups. 


3. Using `groupby` and `.agg()`, find the min and max price for each of the neighbourhood_groups. 


4. Using `groupby`, find the median price for each room type in each neighbourhood_group.


5. Using `groupby` and `.agg()`, find the count, min, max, mean, median, and std of the prices for each room type in each neighbourhood_group.

```python
# 1. Using `groupby`, count how many listings are in each neighbourhood_group.

df.head()
gb = df.groupby("neighbourhood_group")
gb.size() # .size() agg returns size of each group
```

```python
# 2. Using `groupby`, find the mean price for each of the neighbourhood_groups. 

df.head()
gb = df.groupby("neighbourhood_group")

gb["price"].mean()
```

```python
# 2.5. Using `groupby`, find the mean price for each room_type. 
df.head()
gb = df.groupby("room_type")

gb["price"].mean() #.mean() agg returns mean of price for each group
```

```python
# 3. Using `groupby` and `.agg()`, find the min and max price for each of the neighbourhood_groups. 

df.head()
gb = df.groupby("neighbourhood_group")

gb["price"].agg(["min", "max"]) # .agg() function allows for multiple agg at the same time
```

```python
# 4. Using `groupby`, find the mean price for each room_type in each neighbourhood_group.

df.head()
groups = ["neighbourhood_group", "room_type"]
gb = df.groupby(groups) # alternative is to explicitly write group condition in .groupby() function

gb["price"].mean()
```

```python
# 5. Using `groupby` and `.agg()`, find the count, min, max, mean, median, and std of the prices 
# for each room type in each neighbourhood_group.

df.head()
groups = ["neighbourhood_group", "room_type"]
gb = df.groupby(groups) # multi group using .groupby() function

gb["price"].agg(["min", "max", "mean", "median", "std"]) # .agg() for multi aggregation
```
